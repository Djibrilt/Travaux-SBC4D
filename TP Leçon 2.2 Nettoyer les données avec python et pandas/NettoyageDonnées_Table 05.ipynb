```python
# importation des modules

import pandas as pd
import numpy as np

# Importation du fichier source

source = "HDR 2007-2008 Table 05.xlsx"
df = pd.read_excel(source)


# Affichage des 12 premières lignes
df.head(12)
```

```python

# La ligne des variables commence probabelement à la ligne 10
# Repositionner la ligne d'en-tête
df = pd.read_excel(source, header=10)

# Supprimez toutes les colonnes (axe=1) sans valeurs
df = df.dropna(how="all", axis=1)

# Supprimer toutes les lignes (axe=0) sans valeurs
df = df.dropna(how="all", axis=0).reset_index(drop=True)

# Affichons les 5 premières lignes
df.head(5)
```

```python
# Determinons les noms des différentes variables dans les colonnes 
columns = []
for c in df.columns:
    # Pour chaque colonne obtenir les valeurs des 2 premières lignes
    columns.append([str(v)  for v in df[c][:2] if pd.notnull(v)])
columns
```

```python
# Nommons les colonnes
columns = [
    "HDI rank",
    "Country",
    "Total population(millions) 1975",
    "Reference 1",
    "Total population(millions) 2005",
    "Reference 2",
    "Total population(millions) 2015b",
    "Reference 3",
    "Annual population growth rate(%) 1975-2005",
    "Reference 4",
    "Annual population growth rate(%) 2005-15b",
    "Reference 5",
    "Urban population a (% of total) 1975",
    "Urban population a (% of total) 2005",
    "Urban population a (% of total) 2015b",
    "Population under age 15(% of total) 2005",
    "Population under age 15(% of total) 2015b",
    "Population aged 65 and older(% of total) 2005",
    "Population aged 65 and older(% of total) 2015b",
    "Total fertility rate(births per woman) 1970-75c",
    "Total fertility rate(births per woman) 2000-05c"    
]

# Réinitialiser les colonnes des cadres de données
df.columns = columns

# Supprimer les lignes qui contenaient les informations dont nous n'avons plus besoin
# et réinitialiser l'index
df = df.drop(np.arange(2), axis=0).reset_index(drop=True)
df.head()
```

```python
# verifions les dernières lignes de notre fichier
df.tail(30)
```

```python
# Supression des notes de bas de page
df = df.drop(np.arange(180,len(df)), axis=0).reset_index(drop=True)
```

```python
# Catégorisation des HDI
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df = df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)


df.loc[:, "HDI category"] = np.where(df["HDI rank"] < 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI rank"] < 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI rank", inplace=True)
df.head()

# Sauvegarder le fichier de travail
destination = "HDRN 2007-2008 Table 05.csv"
df.to_csv(destination, index=False)
```
